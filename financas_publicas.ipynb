```python
import matplotlib.pyplot as plt

# Dados
categorias = ['Impostos', 'Contribuições', 'Patrimonial', 'Serviços', 'Transferências Correntes', 'Outras Receitas Correntes', 'Amortização de Empréstimos', 'Transferências de Capital']
var_2021_2020 = [20.78, 13.52, 476.18, 21.27, 17.23, 11.77, 80.14, 253.91]
var_2022_2021 = [35.60, 6.48, 392.30, 19.17, 20.85, 24.83, -49.88, 53.54]

# Configurações do gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Posição das barras no eixo X
x = range(len(categorias))

# Plotagem das barras para 2021/2020
plt.bar(x, var_2021_2020, width=0.4, label='2021/2020')

# Plotagem das barras para 2022/2021
plt.bar(x, var_2022_2021, width=0.4, label='2022/2021', bottom=var_2021_2020)

# Rótulos e legendas
plt.xlabel('Categorias de Receita')
plt.ylabel('Variação Percentual (%)')
plt.title('Análise Horizontal das Receitas Públicas (2020-2022)')
plt.xticks(x, categorias, rotation=45, ha='right')
plt.legend()

# Exibição do gráfico
plt.tight_layout()
plt.show()
```
